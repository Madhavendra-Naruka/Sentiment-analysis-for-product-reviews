# 1. Data Collection

I have downloaded Amazon Health and Personal Care Products data, which consists of approximately 500,000 rows and 4.9 million data points.


```python
# Load the JSON file
file_path = 'Health_and_Personal_Care.jsonl'
df = pd.read_json(file_path, lines=True)
print("Shape : ",df.shape)
print("Size : ",df.size)
df=df.sample(10000)
display(df.info())
display(df.head())
```

## Data Preprocessing:

1. **Text Processing:**
   - Combined `title` and `text` columns into `review`.
   - Applied preprocessing (`preprocess_text`) to tokenize and clean text data.

2. **Timestamp Conversion:**
   - Converted `timestamp` column to datetime format.

3. **Sentiment Analysis:**
   - Calculated polarity and subjectivity scores using TextBlob.

4. **Data Cleaning:**
   - Removed unnecessary columns (`title`, `text`, `images`, `asin`, `parent_asin`, `user_id`).

5. **Exporting Data:**
   - Saved processed data to `sample.csv`.

6. **Data Summary:**
   - Displayed DataFrame information.

```python
# Taking out required columns
df['review']=df['title']+" "+df["text"]
tokens=df['review'].apply(preprocess_text)
df['review'] = tokens.apply(lambda x: (" ".join(x)))
df['review'] = df['review'].astype(str)
df['review'] = df['review'].fillna('')
# Convert the timestamp column to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Calculate polarity and subjectivity
df['polarity'] = df['review'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['subjectivity'] = df['review'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# Dropping extra columns
df.drop([1,'title','text','images','asin','parent_asin','user_id'],axis=1,inplace=True)

display(df.head())
df.to_csv("sample.csv", index=False, na_rep='',float_format='%.6f')
display(tokens.head())
display(df.info())
```
